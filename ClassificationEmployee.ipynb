```python
x=df.drop(['Attrition','id'],axis=1)
y=df[['Attrition']]
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
from sklearn.ensemble import RandomForestClassifier
r = RandomForestClassifier(n_estimators = 30, oob_score = True, max_features = 5, max_depth = 30)
r.fit(x, y)
y_pred = r.predict(x)
```

```python
r.score(x, y)
```
